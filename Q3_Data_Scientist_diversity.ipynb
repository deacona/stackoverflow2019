## Data Preparation

```python
df = responses.copy()
df.shape
```

### Labelling Python devs

1=Uses Python, 0=Does not use Python

```python
label = "Data scientist or machine learning specialist" 
# label = "Python"
field = "DevType"
# field = "LanguageWorkedWith"
df[label] = df[field].str.contains(label) \
    .fillna(False).astype(int)
df.dropna(subset=[field], axis=0, inplace=True)

# df["Data Scientist"] = df.DevType.str.contains("Data scientist or machine learning specialist") \
#     .fillna(False).astype(int)
# df.dropna(subset=["DevType"], axis=0, inplace=True)

df[label].value_counts()
```

### Feature selection

We are interested in Gender, Sexuality and Ethnicity

```python
features = [label, "Gender", "Sexuality", "Ethnicity"]
df = df[features]
df.shape
```

### Missing values

Remove/Impute/Work around

```python
df.dropna(subset=features, axis=0, inplace=True)
df.shape
```

### Create feature

Combine our fields into a single "Diversity" feature

```python
df["Diversity"] = df["Gender"] + "~" + df["Sexuality"] + "~" + df["Ethnicity"]
df["Diversity"].value_counts()[:30].plot(kind="bar")
```

```python
df["Diversity score"] = (
    (df.Gender != "Man").astype(int) \
    + (df.Sexuality != "Straight / Heterosexual").astype(int) \
    + (df.Ethnicity != "White or of European descent").astype(int)
) #/ 3
#     * df.Sexuality.str.contains("Straight / Heterosexual")
#     * df.Ethnicity.str.contains("Straight / Heterosexual")
df["Diversity score"].value_counts()
```

## Modelling & Evaluation

```python
print("{0} Diversity types are represented amongst {1} Devs, as opposed to {2} amongst other Devs".format(
    len(df[df[label] == 1].Diversity.unique()) #/ len(df[df[label] == 1])
    , label
    , len(df[df[label] == 0].Diversity.unique()) #/ len(df[df[label] == 0])
))
```

```python
(len(df[df[label] == 1].Diversity.unique()) / len(df[df[label] == 1])) , (len(df[df[label] == 0].Diversity.unique()) / len(df[df[label] == 0]))
```

```python
(len(df[df[label] == 1].Diversity.unique()) / len(df.Diversity.unique())) , (len(df[df[label] == 0].Diversity.unique()) / len(df.Diversity.unique()))
```

```python
groups = []
for i in [1, 0]:
    values = df[df[label] == i]["Diversity score"].value_counts()
#     print(type(values))
    if i == 1:
        name = label
    else:
        name = "Not " + label
    values.rename(name, inplace=True)
#     values.column = i
    groups.append(values)
    
prop = pd.concat(groups, axis=1)
prop = prop.div(prop.sum(1), axis=0)
prop
```

```python
prop.plot(kind="barh", stacked=True, title="Proportion of Developers with different Diversity scores")
```

```python
groups = []
for i in [1, 0]:
    values = df[df[label] == i]["Diversity score"].value_counts() / len(df[df[label] == i])
#     print(type(values))
    if i == 1:
        name = label
    else:
        name = "Not " + label
    values.rename(name, inplace=True)
#     values.column = i
    groups.append(values)
    
prop = pd.concat(groups, axis=1)
prop
```

```python
prop.plot(kind="bar", title="Proportion of Developers with different Diversity scores")
```
