## Data Preparation

```python
df = responses.copy()
df.shape
```

### Labelling developers

1=Belongs to our subgroup of interest, 0=Does not belong to it

```python
label = "Data scientist or machine learning specialist"  # label = "Python"
field = "DevType"  # field = "LanguageWorkedWith"

df[label] = df[field].str.contains(label).fillna(False).astype(int)
df.dropna(subset=[field], axis=0, inplace=True)

df[label].value_counts()
```

### Feature selection

We are interested in Gender, Sexuality and Ethnicity

```python
features = [label, "Gender", "Sexuality", "Ethnicity"]
df = df[features]
df.shape
```

### Missing values

Remove/Impute/Work around

```python
df.dropna(subset=features, axis=0, inplace=True)
df.shape
```

### Create features

Combine our fields into a single "Diversity_group" feature

Also score based on number of fields which differ from base group (white hetrosexual men)

```python
df["Diversity_group"] = df["Gender"] + "~" + df["Sexuality"] + "~" + df["Ethnicity"]
df["Diversity_group"].value_counts()[:30].plot(kind="bar")
```

```python
df["Diversity_score"] = (
    (df.Gender != "Man").astype(int)
    + (df.Sexuality != "Straight / Heterosexual").astype(int)
    + (df.Ethnicity != "White or of European descent").astype(int)
)

df["Diversity_score"].value_counts()
```

## Modelling & Evaluation

```python
print(
    "{0} Diversity groups are represented amongst {1} Devs, as opposed to {2} amongst other Devs".format(
        len(df[df[label] == 1].Diversity_group.unique()),  # / len(df[df[label] == 1])
        label,
        len(df[df[label] == 0].Diversity_group.unique()),  # / len(df[df[label] == 0])
    )
)
```

```python
groups = []
for i in [1, 0]:
    values = df[df[label] == i]["Diversity_score"].value_counts()
    #     print(type(values))
    if i == 1:
        name = label
    else:
        name = "Not " + label
    values.rename(name, inplace=True)
    #     values.column = i
    groups.append(values)

prop = pd.concat(groups, axis=1)
prop = prop.div(prop.sum(1), axis=0)
prop
```

```python
prop.plot(
    kind="barh",
    stacked=True,
    title="Proportion of Developers with different Diversity scores",
)
```

```python
groups = []
for i in [1, 0]:
    values = df[df[label] == i]["Diversity_score"].value_counts() / len(
        df[df[label] == i]
    )
    #     print(type(values))
    if i == 1:
        name = label
    else:
        name = "Not " + label
    values.rename(name, inplace=True)
    #     values.column = i
    groups.append(values)

prop = pd.concat(groups, axis=1)
prop
```

```python
prop.plot(kind="bar", title="Proportion of Developers with different Diversity scores")
```

```python
for i in [1, 0]:
    if i == 1:
        prefix = ""
    else:
        prefix = "Not "

    ratio = len(df[df[label] == i]) / len(df[df[label] == i].Diversity_group.unique())

    print(
        "{0}{1}: 1 in {2:0.1f} belong to different diversity grouping".format(
            prefix, label, ratio,
        )
    )
```

## Deployment
